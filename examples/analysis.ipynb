```python
import os
path_input = "/data/output/rocs/results_01_individual.h5"
path_input
```

```python
import pandas as pd
```

```python
df = pd.read_hdf(path_input)
# df.index = df.index.set_names(['path', 'model', 'run id'])
```

```python
df_mean = df.groupby(level=["path","model"], group_keys=False).mean()
df_mean['run id'] = 'mean'
df_mean.set_index('run id', append=True, inplace=True)

df_std = df.groupby(level=["path","model"], group_keys=False).std()
df_std['run id'] = 'std'
df_std.set_index('run id', append=True, inplace=True)

df_global = df.groupby(level=["model"], group_keys=False).mean()
df_global = df_global.reset_index()
df_global['run id'] = 'mean'
df_global['path'] = 'all'
df_global.set_index(['path', 'model', 'run id'], append=False, inplace=True)


df_final = pd.concat([
    df, 
    df_mean,
    df_std,
    df_global
    ]
    )

def split_path(row):
    path = row.name[0].strip()
    row['id'] = path
    row['class'] = path
    if(len(path) > 3):
        row['id'] = path.split(' ')[-1] 
        row['class'] = path.split(' ')[-2]
    return row
df_final = df_final.apply(split_path, axis=1)

df_final = df_final.reset_index()

df_final = df_final.drop(columns='path')
df_final = df_final.set_index(['class', 'id', 'model', 'run id'])

df_final
```
